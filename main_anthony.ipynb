```python
import ultralytics
from ultralytics import YOLO
ultralytics.checks()

model = YOLO("runs/segment/train/weights/best.pt")
```

```python
results = model.predict("test1.jpg", imgsz=640)
```

```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.patches as patches
# bg_folder = "backgrounds"
# cards_path = "pokemon_cards"
# final_img, bboxes = create_image_with_random_cards(bg_folder, cards_path)
final_img = results[0].orig_img
# Display the final image using OpenCV's imshow
cv2.imshow('Final Image with Cards', final_img)
cv2.waitKey(0)  # Wait for a key press
cv2.destroyAllWindows()  # Close the window when done

bboxes = results[0].boxes.xyxy.cpu().numpy()


final_img_rgb = cv2.cvtColor(final_img, cv2.COLOR_BGR2RGB)
# Plot the image
plt.figure(figsize=(8, 6))
plt.imshow(final_img_rgb)
plt.axis('on')
print(bboxes)
# Plot and annotate the bounding boxes
for bbox in bboxes:
    print(bbox)
    x_min, y_min, x_max, y_max = bbox
    # Create a rectangle patch for the bounding box
    rect = patches.Rectangle(
        (x_min, y_min),  # Bottom-left corner
        x_max - x_min,  # Width
        y_max - y_min,  # Height
        linewidth=2,
        edgecolor='red',
        facecolor='none'
    )
    plt.gca().add_patch(rect)
    # Annotate the bounding box
    plt.text(
        x_min, y_min - 10, f'({x_min}, {y_min}, {x_max}, {y_max})',
        color='yellow',
        fontsize=10,
        bbox=dict(facecolor='black', alpha=0.5)
    )

# Display the result
plt.title("Final Image with Bounding Boxes")
plt.show()
```

```python
from card_scanner.perspective_transform import perspective_transform

img = results[0].orig_img
card_images = []

for mask in results[0].masks:
    mask = mask.data.cpu().numpy().squeeze()
    card_image = perspective_transform(img, mask, False)
    if card_image is not None:
        card_image = cv2.cvtColor(card_image, cv2.COLOR_BGR2RGB)
        card_images.append(card_image)
```

```python
len(card_images)
```

```python
import matplotlib.pyplot as plt

# Create a 2x4 grid layout
fig, axes = plt.subplots(2, 4, figsize=(12, 6))  # Adjust figsize as needed

# Flatten the axes for easy iteration
axes = axes.flatten()

# Plot each image in its corresponding subplot
for i, ax in enumerate(axes):
    if i < len(card_images):
        ax.imshow(card_images[i])  # Show the image
        ax.axis('off')            # Hide axes
    else:
        ax.axis('off')            # Turn off empty plots

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```
